```python
# Cell 1: Install dependencies
!pip install opencv-python-headless numpy matplotlib

# Cell 2: Import libraries & define functions
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt

def load_image(path):
    img = cv2.imdecode(np.frombuffer(open(path, 'rb').read(), np.uint8), cv2.IMREAD_COLOR)
    if img is None:
        raise ValueError(f"{path} is not a valid image!")
    return img

def lsb_change_stats(original, encoded):
    # Only blue channel for simplicity (you can change to all channels if needed)
    orig_lsb = (original[:,:,0] & 1).flatten()
    enc_lsb = (encoded[:,:,0] & 1).flatten()
    changed = np.sum(orig_lsb != enc_lsb)
    total = orig_lsb.size
    percent = 100.0 * changed / total if total else 0
    return changed, total, percent

def plot_lsb_diff_map(original, encoded):
    diff = ((original[:,:,0] & 1) != (encoded[:,:,0] & 1)).astype(np.uint8) * 255
    plt.figure(figsize=(6,6))
    plt.imshow(diff, cmap='hot')
    plt.title('LSB Change Map (Blue Channel)')
    plt.axis('off')
    plt.show()

def analyze_images(original_path, encoded_path, threshold=0.5):
    original = load_image(original_path)
    encoded = load_image(encoded_path)
    if original.shape != encoded.shape:
        print("Error: Images must be the same size and shape!")
        return
    changed, total, percent = lsb_change_stats(original, encoded)
    print(f"Total pixels compared: {total}")
    print(f"Number of LSB changes: {changed}")
    print(f"Percentage of LSB changed: {percent:.4f}%")
    if percent > threshold:
        print("\n\033[92mResult: Encoded image contains hidden data (LSB pattern significantly changed)\033[0m")
    else:
        print("\n\033[91mResult: No significant hidden data found in LSB (changes are minimal)\033[0m")
    plot_lsb_diff_map(original, encoded)

# Cell 3: Upload original and encoded images
print("Upload the ORIGINAL image first:")
uploaded = files.upload()
orig_file = list(uploaded.keys())[0]

print("Now upload the ENCODED (stego) image:")
uploaded = files.upload()
enc_file = list(uploaded.keys())[0]

print(f"\nAnalyzing {orig_file} vs {enc_file} ...")
analyze_images(orig_file, enc_file, threshold=0.5)  # Threshold can be tuned
```
