```python
from PIL import Image
from google.colab import files

# --- LSB extraction ---
def extract_lsb(img: Image.Image):
    if img.mode != "RGB":
        img = img.convert("RGB")
    w,h = img.size
    px = img.load()
    total_pixels = w*h
    pixels_with_data = 0
    bits = []

    for y in range(h):
        for x in range(w):
            r,g,b = px[x,y]
            lsb_bits = [r&1, g&1, b&1]
            if any(lsb_bits): pixels_with_data += 1
            bits.extend(lsb_bits)

    used_bits = len(bits)
    capacity_pct = 100*used_bits/(total_pixels*3)

    # ASCII readability check
    byte_data=[]
    for i in range(0,len(bits)-len(bits)%8,8):
        byte=0
        for b in bits[i:i+8]: byte=(byte<<1)|b
        byte_data.append(byte)
    ascii_ratio = 100*sum(32<=b<=126 for b in byte_data)/len(byte_data) if byte_data else 0

    return {
        "total_pixels": total_pixels,
        "pixels_with_data": pixels_with_data,
        "used_bits": used_bits,
        "capacity_pct": capacity_pct,
        "ascii_ratio": ascii_ratio,
        "byte_data": bytes(byte_data)
    }

# --- Report generation ---
def generate_report(stats):
    print("### Image Steganography Analysis Report\n")
    print("### 1. Visual Attack Analysis")
    print(f"- Total Pixels: {stats['total_pixels']}")
    print(f"- Pixels with LSB Data: {stats['pixels_with_data']}")
    visual_ratio = stats['pixels_with_data']/stats['total_pixels']*100
    print(f"- Approx. % of pixels affected: {visual_ratio:.2f}%")
    if visual_ratio<5:
        print("- Conclusion: Likely imperceptible by human eye.")
    else:
        print("- Conclusion: Some visual cues may exist.")

    print("\n### 2. Statistical Attack Analysis")
    bit_ratio = stats['used_bits']/(stats['total_pixels']*3)
    print(f"- LSB usage ratio: {bit_ratio:.2f}")
    if bit_ratio<0.05:
        print("- Conclusion: Low capacity, difficult to detect statistically.")
    else:
        print("- Conclusion: Higher capacity, may be detectable.")

    print("\n### 3. Capacity Analysis")
    print(f"- Bits used: {stats['used_bits']} (~{stats['capacity_pct']:.2f}% of total image capacity)")

    print("\n### 4. Comparative & Error Analysis")
    print(f"- ASCII-like ratio: {stats['ascii_ratio']:.2f}%")
    if stats['ascii_ratio']>50:
        print("- Data likely human-readable text.")
    else:
        print("- Data likely binary or encrypted.")
    print("- Partial recovery possible if full decode fails.")

    print("\n### 5. Extracted Preview (first 500 chars if text)")
    try:
        text = stats['byte_data'].decode("utf-8", errors="replace")
        print(text[:500])
    except:
        print("Cannot decode as UTF-8; data may be binary.")

# --- Main ---
print("üì§ Upload PNG/BMP image:")
uploaded = files.upload()
if uploaded:
    img_path = list(uploaded.keys())[0]
    img = Image.open(img_path)
    stats = extract_lsb(img)
    generate_report(stats)
```

```python
import math
from collections import Counter
from google.colab import files
import matplotlib.pyplot as plt

# --- Entropy Calculation ---
def entropy(data):
    if not data: return 0
    counter = Counter(data)
    length = len(data)
    return -sum((count/length)*math.log2(count/length) for count in counter.values())

# --- LSB Analysis ---
def lsb_analysis(data, bits_per_byte=1):
    bits = []
    for b in data:
        for i in range(bits_per_byte):
            bits.append((b >> i) & 1)
    ones_ratio = sum(bits)/len(bits)*100 if bits else 0
    return bits, ones_ratio

# --- ASCII / Unicode Ratio ---
def ascii_ratio(data):
    if not data: return 0
    return sum(32<=b<=126 for b in data)/len(data)*100

# --- Bit Distribution Plot ---
def plot_bit_distribution(bits, filename="bit_dist.png"):
    counts = [0,0]
    for b in bits:
        counts[b] += 1
    plt.bar([0,1], counts, color=['blue','red'])
    plt.xticks([0,1], ["0","1"])
    plt.ylabel("Count")
    plt.title("Bit Distribution (LSB Analysis)")
    plt.savefig(filename)
    plt.close()
    return filename

# --- PBT Detection ---
def pbt_detection(file_path):
    with open(file_path,"rb") as f:
        content = f.read()

    file_entropy = entropy(content)
    bits, lsb_ratio = lsb_analysis(content, bits_per_byte=1)
    ascii_r = ascii_ratio(content)

    print(f"üìÅ File: {file_path}")
    print(f"- Size: {len(content)} bytes")
    print(f"- Entropy: {file_entropy:.2f} bits/byte")
    print(f"- LSB 1s ratio: {lsb_ratio:.2f}%")
    print(f"- ASCII-like ratio: {ascii_r:.2f}%")

    # Heuristic
    if lsb_ratio > 45 and ascii_r < 50:
        print("‚ö†Ô∏è Possible hidden data detected (high LSB usage + low ASCII)")
    elif file_entropy > 7.5:
        print("‚ö†Ô∏è High entropy detected ‚Üí file may contain encrypted/stego data")
    else:
        print("‚úÖ No obvious hidden data detected")

    # Show bit distribution
    plot_file = plot_bit_distribution(bits)
    print(f"üìä Bit distribution chart saved as {plot_file}")

    # Partial text preview if ASCII
    if ascii_r > 30:
        preview = ''.join(chr(b) if 32<=b<=126 else '.' for b in content[:500])
        print(f"\nüìù ASCII Preview (first 500 chars):\n{preview}")

# --- Main ---
print("üì§ Upload .pbt/.pyc/.py file for detection:")
uploaded = files.upload()
if uploaded:
    for path in uploaded.keys():
        pbt_detection(path)
```
